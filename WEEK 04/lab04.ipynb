# Lab

Using the flights dataframe answer the following questions

1. When do the first and last flights leave each day?

2. When there is a missing value in `dep_time` then this is an indication of canceled flight. Find the number of cancelled flights for each (`carrier`, `origin`) combination.

3. Which carrier has the worst `dep_delay`s? 

4. Which plane (`tailnum`) has the worst on-time record?

5. For each plane, count the number of flights before the first delay of greater than 1 hour.

```python
import pandas as pd

flights = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/nyc_flights.csv')
```

```python
flights
```

First and Last flight each day

```python
groupby_day = flights \
    .groupby(['month', 'day']) \
    .agg( first_flight = ('dep_time', 'min'), Last_flight = ('dep_time', 'max'))
```

```python
groupby_day
```

number of cancelled flights for each (carrier, origin) combination.

```python
def nan_count(x):
    
    return x.isnull().sum()

```

```python
Carrier_null = flights \
    .groupby(['carrier', 'origin']) \
    .agg( NaN_count = ('dep_delay', nan_count)) 

Carrier_null
```

carrier that has the worst departure delay

```python
worst_delay = flights.groupby('carrier').agg( worst_delay = ('dep_delay', 'max'))

max_worst_delay = worst_delay.loc[(worst_delay['worst_delay'] == worst_delay['worst_delay'].max())]

max_worst_delay
```

plane (tailnum)that has the worst on-time record

```python
ontime_delay = flights.groupby('tailnum').agg(arrival_delay = ('arr_delay', 'max'))

ontime_delay

worst_ontime_flight = ontime_delay.loc[(ontime_delay['arrival_delay'] == ontime_delay['arrival_delay'].max())]

worst_ontime_flight

```

```python
highest_arrival_delay = flights.loc[(flights['arr_delay'] == flights['arr_delay'].max())]
highest_arrival_delay
```

number of flights before the first delay of greater than 1 hour

```python
delay_lessthan_hour = flights.loc[(flights['arr_delay'] < 1)]
delay_lessthan_hour.groupby(['carrier', 'flight']).agg(Less_Than_Hour_Delay = ('arr_delay', 'count'))
```
