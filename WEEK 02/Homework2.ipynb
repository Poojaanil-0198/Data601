````python
import random 
import matplotlib.pyplot as plt

trial_count = 500

def fair_die_simulation(trials):
    """
    returns an array of randomly numbers between 1 and 6. 
    """
    options = [1,2,3,4,5,6]
    simulations = []
    for _ in range(trials):
         simulations.append(random.choice(options))
        
    return simulations

fair_die_simulation(500)
````

````python
import numpy 
def biased_die_simuations(trials, weights=[0.20, 0.10, 0.15, 0.15, 0.15, 0.25]):
    """
    returns an array of randomly numbers between 1 and 6 with the probabilty of assosiated weight. 
    for instance: for weights: [0.25, 0.15, 0.15, 0.15, 0.15, 0.15]
    1 has 0.25 change to be selected, 2 has 0.15 chance and so on...
    """
    
    #options = [1,2,3,4,5,6]
    simulations = []
    for _ in range(trials):
        p=weights
        simulations.append(numpy.random.choice(numpy.arange(1, 7), p=weights ))
    return simulations

biased_die_simuations(500) 
````

````python
from numpy import mean      #mean calculation for fair dice simulation
def fair_mean_lists(trials):
    fair_mean_list = []
    for _ in range (trials):
        fair_mean_list.append(mean(fair_die_simulation(trials)))
    return fair_mean_list
        
fair_mean_lists(500)
````

````python
#mean calculation for biased dice simulation
def biased_mean_lists(trials):
    biased_mean_list = []
    for _ in range (trials):
        biased_mean_list.append(mean(biased_die_simuations(trials)))
    return biased_mean_list
        
biased_mean_lists(500)
````

````python
biased = biased_die_simuations(trial_count)
fair = fair_die_simulation(trial_count)

# biased_mean_list = ...
biased_mean_list = biased_mean_lists(trial_count)
# unbiased_mean_list = ...
unbiased_mean_list = fair_mean_lists(trial_count)
````

````python
plt.figure(figsize =(15, 15))      #create new figure

plt.subplot(221)                   #Fair Count plot
plt.title('fair dice', fontweight="bold", fontsize=30)
x_axis = [1, 2, 3, 4, 5, 6]
count = [fair.count(1),fair.count(2),fair.count(3),fair.count(4),fair.count(5),fair.count(6)]
plt.bar(x_axis, count)
plt.xlabel('dice side')
plt.ylabel('count')

plt.subplot(222)                   #Biased Count plot
plt.title('biased dice', fontweight="bold", fontsize=30)
x_axis = [1, 2, 3, 4, 5, 6]
count = [biased.count(1),biased.count(2),biased.count(3),biased.count(4),biased.count(5),biased.count(6)]
plt.bar(x_axis, count)
plt.xlabel('dice side')
plt.ylabel('count')

plt.subplot(2, 2, (3, 4))           #Comparing Fair and Biased mean 
plt.title('mean comparision', fontweight="bold", fontsize=30)
plt.hist([unbiased_mean_list, biased_mean_list])
labels = ['unbiased', 'biased']
plt.legend(labels)
plt.xlabel('count')
plt.ylabel('mean')
plt.show()
````
