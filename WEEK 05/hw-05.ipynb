```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np 
```

```python
#Checking the number of flights for each carrier  #flights.carrier.value_counts()
Carrier_Count = flights.groupby(['carrier']).agg({'flight' : 'count'})
Carrier_Count
```

```python
plt.bar( x = Carrier_Count.index , height = Carrier_Count.flight)
plt.show()
```

 1. CHECKING WHICH CARRIER HAS THE MAXIMUM FLIGHT ACTIVITIES (CATEGORICAL VARIABLE)

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.histplot(data = flights, x = 'carrier')
```

REPORT: this graph shows that UNITED AIRLINES has the most flights travelling across the states.

2. CHECKING FOR THE MONTH THAT HAD THE MOST FLIGHTS (CONTINOUS VARIABLE)

```python
flights_by_month = flights.groupby(['month']).agg({'flight' : 'count'})
flights_by_month
```

```python
sns.set_style('ticks')
sns.catplot( data = flights, x = 'month', kind = 'count')
```

REPORT :  FEB has the least number of flight travels 

CHECKING FOR THE NUMBER OF DEPARTURE DELAY(mean) FOR DESTINATION AIRPORT

```python
origins1 = flights.origin.value_counts()
origins1
```

```python
#as dep_delay has 8255 null values, I will be dropping the null values 
origin_data = flights.dropna(subset = ['dep_delay', 'arr_delay'])
origins = origin_data.groupby('dest').dep_delay.mean()
origins
```

```python
plt.figure(figsize = (20, 7))
plt.bar(origins.index, origins.values)
plt.xlabel('Destinations')
plt.ylabel('Average Delays')
plt.xticks(rotation = 90)
plt.show()
```

REPORT: destinations CAE - Columbia and TUL - Tulsa have the most delays.

NUMBER OF FLIGHTS DEPARTURING FROM ORIGINS AND DESTINATION

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.histplot(data = origin_data, x = 'origin')
```

```python
sns.set(rc = {'figure.figsize':(20,8)})
sns.histplot(data = origin_data, x = 'dest')
plt.xticks(rotation = 90)
plt.show()
```

REPORT: Most flights departure from the origin EWR which is the Newark Airport and most flights arrive at ATL and ORD,which is Atlanta and Chicago respectively.
