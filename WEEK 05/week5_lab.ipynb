```python
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.scatterplot(data = diamonds, x = "carat", y = "price", hue = "clarity")
```

The pattern shows that as the carat of the diamond increases, the price of the diamond increases along with it. s12 clarity dominates the others 

#### 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
plot = sns.displot(diamonds, x="clarity", y="color", cbar = True)
plot.ax.tick_params(labelrotation = 90)
```

This pattern shows that the diamonds with color G and clarity VS21 are the most common and are high in number.

- Use `carat` variable as as the size of the points in scatter plot in which x-axis is the `depth` and `price` on the y-axis.
