```python
from zipfile import ZipFile
from urllib import request
```

```python
def download_file(file_name, url):
    res = request.urlopen(url)
    with open(file_name, 'wb')as file:
        file.write(res.read())
        
def unzip(file_name, path = './'):
    with ZipFile(file_name, 'r') as zip:
        zip.printdir()
        print('Extracting all the files now....')
        zip.extractall(path = path)
        print('Done!')
        
    
```

```python
file_url = 'https://github.com/msaricaumbc/DS_data/blob/master/bigfile.csv.zip?raw=true'
file_name = 'bigfile.csv.zip'

download_file(file_name, file_url)
unzip(file_name)
```

```python
Occupation = []
with open('bigfile.csv', 'r', newline='') as file:
    age = []
    i = 0
    row_count = 0
    while(True):
        line = file.readline()
        if not line.split():
            break
        data = line.split(',')
        age.append(data[3])
        occupation = line.split(',')[4]
        Occupation.append(occupation)
        #line_array = line.split(',')[:6]
        #age = line.split(',')[3:4]
        #print(line_array)
        row_count +=1
        
        i=i+1
```

Number of records processed

```python
print(row_count)
```

```python
age = [int(x) for x in age]
age.sort()
from matplotlib import pyplot as plt
import numpy as np
plt.figure(figsize = (20, 15))
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.hist(age)
```

```python
frequency = {}
for item in Occupation:
    if item in frequency:
        frequency[item] +=1
    else:
        frequency[item] = 1
occupation_dict= dict(sorted(frequency.items(), key=lambda item: item[1], reverse=True))
occupation_list = list(occupation_dict.items())
occupation_list[0:20]
```
