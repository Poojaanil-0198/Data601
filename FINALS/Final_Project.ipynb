For the current purpose, only the "skin care products" category fits right and is inline with the **Products** dataframe. In the below cell, the dataframe has been filtered to consider the same.

```python
chemicals = chemicals.loc[chemicals['PrimaryCategory'].str.contains("Skin Care Products", case = False, na=False)]
chemicals.head()
```

**NOTE :** The Subcategory section specifies the the function of the skin care products retrieved from the last cell. The names and renamed to match that of dataset 1. For example, the <b>facial creams</b> are renamed as <b>moisturizers</b> for accuracy.

```python
chemicals.SubCategory.value_counts()
```

```python
chemicals['SubCategory'] = chemicals['SubCategory'].replace('Facial Cream','Moisturizer')
chemicals['SubCategory'] = chemicals['SubCategory'].replace('Skin Cleansers','Cleanser')
chemicals['SubCategory'] = chemicals['SubCategory'].replace('Facial Masks','Face mask')
```

**ADDITIONAL RESEARCH :**<ol>
    <li><b>TITANIUM DIOXIDE :</b> Titanium dioxide has been classified as **IARC Group 2B carcinogenic** to humans by Internation Agency for Research on Cancer. It is used in the skin care products for pigmentation and as an opacifier due to its white color and ability to refract light.</li>
    <li><b>BENZOPHENONE : </b>This is generally used to block the UV rays from damaging the other ingredients present in the product. Benzophenone is know to cause **weak hormonal activities** in lab tested animals. It is also known to be **carcinogenic**. Benzophenone and its derivatives can also be the reason for severe **endocrine disruptions and organ system toxicity**.</li>
</ol>
